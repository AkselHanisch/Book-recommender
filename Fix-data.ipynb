```python
# Import Required Libraries
import pandas as pd

# Load Datasets
ratings = pd.read_csv('data/ratings.csv')

# Display the first few rows of the ratings dataset
print("Ratings Data:")
print(ratings.head())

# Count the number of reviews for each book
book_review_counts = ratings.groupby('ISBN')['Book-Rating'].apply(lambda x: (x > 0).sum())

# Display the number of books with and without valid reviews
total_books = len(book_review_counts)
books_with_valid_reviews = (book_review_counts > 0).sum()
books_without_valid_reviews = total_books - books_with_valid_reviews

print(f"\nTotal number of books: {total_books}")
print(f"Number of books with valid reviews: {books_with_valid_reviews}")
print(f"Number of books without valid reviews: {books_without_valid_reviews}")
```

```python
# Import Required Libraries
import pandas as pd

# Load Datasets
ratings = pd.read_csv('data/ratings.csv')
books = pd.read_csv('data/books.csv')

# Display the first few rows of the datasets
print("Ratings Data:")
print(ratings.head())

print("\nBooks Data:")
print(books.head())

# Filter out books without valid ratings (ratings equal to 0)
valid_ratings = ratings[ratings['Book-Rating'] > 0]

# Get the list of ISBNs with valid ratings
valid_isbns = valid_ratings['ISBN'].unique()

# Filter the books dataset to include only books with valid ratings
filtered_books = books[books['ISBN'].isin(valid_isbns)]

# Display the first few rows of the filtered books dataset
print("\nFiltered Books Data:")
print(filtered_books.head())

# Save the filtered books dataset to a new CSV file
filtered_books.to_csv('data/filtered_books.csv', index=False)

# Compare the number of books in the original and filtered datasets
original_books_count = len(books)
filtered_books_count = len(filtered_books)

print(f"\nTotal number of books in the original dataset: {original_books_count}")
print(f"Total number of books in the filtered dataset: {filtered_books_count}")
```
